```python
fraud_df = pd.read_csv('/kaggle/input/credit-card-fraud-detection-dataset-2023/creditcard_2023.csv', header = 0)
fraud_df.head()
```

```python
fraud_df.info()
fraud_df.describe()
```

```python
exclude_column = ['id', 'Class']

# Calculate the mean and standard deviation for each column (excluding the exclude_column)
mean_std = fraud_df.drop(exclude_column, axis=1).agg(['mean', 'std'])

# Normalize the columns (excluding the exclude_column)
normalized_df = (fraud_df.drop(exclude_column, axis=1) - mean_std.loc['mean']) / mean_std.loc['std']

# Combine the normalized columns with the exclude_column
normalized_df[exclude_column] = fraud_df[exclude_column]

# Display the normalized DataFrame
normalized_df.describe()
```

Mean close to zero, std close to 1.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# X contains features (29 features)
X = normalized_df.drop(exclude_column, axis =1)
# y contains your binary labels (0 or 1)
y = normalized_df['Class']

seed = 2023
# Split the dataset into training and testing sets
# test size is 0.02 * 568630 = 11373
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.02, random_state=seed)

# Standardize features (mean=0, std=1)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create a Multi-Layer Perceptron (MLP) Classifier
mlp = MLPClassifier(hidden_layer_sizes=(128, 64), activation='relu', solver='adam', max_iter=200, random_state=seed)

# Train the model
mlp.fit(X_train, y_train)
```

```python
# Make predictions on the train set
y_pred = mlp.predict(X_train)

# Evaluate the model- Train
accuracy = accuracy_score(y_train, y_pred)
precision = precision_score(y_train, y_pred)
recall = recall_score(y_train, y_pred)
f1 = f1_score(y_train, y_pred)
roc_auc = roc_auc_score(y_train, mlp.predict_proba(X_train)[:, 1])

# Print evaluation metrics
print('-----Train------')
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")

# Make predictions on the test set
y_pred = mlp.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, mlp.predict_proba(X_test)[:, 1])

# Print evaluation metrics
print('-----Test------')
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score

# Create a Keras model with dropout layers
model = Sequential()
model.add(Dense(128, input_dim=29, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer with a 20% dropout rate
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer with a 20% dropout rate
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0001), metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=256, validation_data=(X_test, y_test))
```

```python
# Make predictions on the test set
y_pred_test = (model.predict(X_test) > 0.5).astype(int)

# Make predictions on the train set
y_pred_train = (model.predict(X_train) > 0.5).astype(int)

# Evaluate the model
accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {accuracy_train:.4f}")
print(f"Test Accuracy: {accuracy_test:.4f}")
```
